```python
import sqlite3
conn=sqlite3.connect('my_database.db')
```

## 1)  Display all the details of movies created by directors Christopher and Matt Reeves.

```python
result1=conn.execute("SELECT * FROM data1 WHERE Director1 IN ('Christopher Nolan ', 'Matt Reeves ');")
for i in result1:
    print(i)
```

##  2) Display all the details of movies with a duration of 140 minutes to 190 minutes.

```python
result2=conn.execute("SELECT * FROM data1 WHERE Duration >= 140 AND Duration <= 190;")
for i in result2:
    print(i)
```

## 3) Display all details of movies with ratings above 7 in ascending order.

```python
result3=conn.execute("SELECT * FROM data1 WHERE Rating > 7 ORDER BY Rating ASC;")
for i in result3:
    print(i)
```

## 4) Display all movie names in descending order.

```python
result4=conn.execute("SELECT MovieName FROM data1 ORDER BY MovieName DESC; ")
for i in result4:
    print(i)
```

## 5) Display movie name starts with ‘P’ and their rating is greater than 7.

```python
result5=conn.execute("SELECT MovieName FROM data1 WHERE MovieName LIKE 'P%' AND Rating > 7;")
for i in result5:
    print(i)
```

## 1. Display all details from both tables where movie names are same.

```python
result11=conn.execute("SELECT * FROM data1 JOIN data2 ON data1.MovieName=data2.MovieName;")
for i in result11:
    print(i)
```

## 2. Display all movie names, Director, ratings, and gross collection where the genre is action.

```python
result12=conn.execute("SELECT data1.MovieName, data1.Director1, data1.Rating, data2.GrossCollection, data2.Genre1 FROM data1, data2 WHERE (data2.Genre1='Action' ) AND data1.MovieName=data2.MovieName;")
for i in result12:
    print(i)
```

## 3. Display all the details from both tables with the highest gross collection.

```python
result13=conn.execute("SELECT*FROM data1, data2 WHERE data2.GrossCollection=(SELECT MAX(GrossCollection) FROM data2) AND data1.MovieName=data2.MovieName;")
for i in result13:
    print(i)
```

## 4. Display all the details from both the tables with the highest ratings.

```python
result14=conn.execute("SELECT*FROM data1, data2 WHERE data1.Rating=(SELECT MAX(Rating) FROM data1) AND data1.MovieName=data2.MovieName;")
for i in result14:
    print(i)
```

## 5. Display all the details from both tables with the lowest gross collection and lowest ratings.

```python
result15=conn.execute("SELECT*FROM data1,data2 WHERE(data2.GrossCollection=(SELECT min(GrossCollection) FROM data2) OR data1.Rating=(SELECT min(Rating) FROM data1)) AND data1.MovieName=data2.MovieName;")
for i in result15:
    print(i)
```
